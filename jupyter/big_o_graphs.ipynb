```python
import numpy as np  
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import scipy.special
```

```python
def o_of_log_n(x_range):
    x = np.array(x_range)
    y = np.log(x)
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n)')
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_n(x_range):
    x = np.array(x_range)
    y = x
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n)')
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_n_log_n(x_range):
    x = np.array(x_range)
    y = x * np.log(x)
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n)')
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_1(x_range):
    x = np.array(x_range)
    y = np.array([ 1 for _ in x_range ])
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n)')
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_n_squared(x_range):
    x = np.array(x_range)
    y = x**2
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n)')
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_2_pow_n(x_range):
    x = np.array(x_range)
    y = 2**x
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n)')
    ax = plt.figure().gca()
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_n_fac(x_range):
    x = np.array(x_range)
    y = scipy.special.factorial(x)
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n)')
    plt.plot(x, y)
    plt.show()
```

```python
o_of_1(range(0, 101))
```

```python
o_of_log_n(range(1, 101))
```

```python
o_of_n(range(0, 101))
```

```python
o_of_n_log_n(range(1, 11))
```

```python
o_of_n_squared(range(0, 101))
```

```python
o_of_2_pow_n(range(0, 21))
```

```python
o_of_n_fac(range(0, 6))
```
