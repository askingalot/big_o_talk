```python
import numpy as np  
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import scipy.special
```

```python
def o_of_log_n(x_range):
    x = np.array(x_range)
    y = np.log2(x)
    plt.xlabel('Size of input - n')
    plt.ylabel('O(log n)')
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_n(x_range):
    x = np.array(x_range)
    y = x
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n)')
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_n_log_n(x_range):
    x = np.array(x_range)
    y = x * np.log2(x)
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n * log n)')
    #plt.ylim((0, 400))
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_1(x_range):
    x = np.array(x_range)
    y = np.array([ 1 for _ in x_range ])
    plt.xlabel('Size of input - n')
    plt.ylabel('O(1)')
    plt.ylim((0,5))
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_n_squared(x_range):
    x = np.array(x_range)
    y = x**2
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n^2)')
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_2_pow_n(x_range):
    x = np.array(x_range)
    y = 2**x
    plt.xlabel('Size of input - n')
    plt.ylabel('O(2^n)')
    plt.xticks(np.arange(0, 21, 5))
    plt.plot(x, y)
    plt.show()
```

```python
def o_of_n_fac(x_range):
    x = np.array(x_range)
    y = scipy.special.factorial(x)
    plt.xlabel('Size of input - n')
    plt.ylabel('O(n!)')
    plt.plot(x, y)
    plt.show()
```
